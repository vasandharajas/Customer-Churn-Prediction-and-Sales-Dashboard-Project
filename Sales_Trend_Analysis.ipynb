# Sales Trend Analysis to Track monthly, quarterly, and yearly sales growth.And Identify top-performing products.Correlate churn trends with revenue impact.

```python
## Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#### Load Data
Data_path = r"feature_engineering_preprocessed_data.csv"
df = pd.read_csv(Data_path)
```

```python
#### To Prepare The Date Columns
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create time period columns
df['Month'] = df['Order Date'].dt.to_period('M')
df['Quarter'] = df['Order Date'].dt.to_period('Q')
df['Year'] = df['Order Date'].dt.year
```

```python
#### To Visualizing the Monthly Sales Trend
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10, 5))
monthly_sales.plot(kind='line', marker='o', color='teal')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#### To Visualizing the Quarterly Sales Trend
quarterly_sales = df.groupby('Quarter')['Sales'].sum()

plt.figure(figsize=(10, 5))
quarterly_sales.plot(kind='bar', color='coral')
plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
#### To Visualizing the Yearly Sales Trend
yearly_sales = df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(8, 5))
yearly_sales.plot(kind='bar', color='slateblue')
plt.title("Yearly Sales Growth")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
####  To Visualizing the Top-Performing Products
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
#### To Save the Data of Monthly, Quarterly, Yearly CSVs File
### Save Path
Save_Path = r"your datas save path"

monthly_sales.to_csv(os.path.join(Save_Path, 'monthly_sales.csv'))
quarterly_sales.to_csv(os.path.join(Save_Path, 'quarterly_sales.csv'))
yearly_sales.to_csv(os.path.join(Save_Path, 'yearly_sales.csv'))
top_products.to_csv(os.path.join(Save_Path, 'top_products.csv'))

print("Sales trend analysis completed and saved successfully.")
```
